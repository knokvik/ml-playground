```python
import pandas as pd

data = {
    "department" : ["IT","IT","HR","HR"],
    "gender" : ["Male","Female","Male","Female"],
    "salary" : [50000,60000,40000,45000]
}

df = pd.DataFrame(data)

pd.pivot_table(
    df,
    values=df.columns[2],
    index=df.columns[0],
    aggfunc='mean'
)
```

```python
pd.pivot_table(
    df,
    values=df.columns[2],
    index=df.columns[0],
    columns=df.columns[1],
    aggfunc='mean'
)
```

```python
pd.crosstab(
    df[df.columns[0]],
    df[df.columns[2]],
    normalize='index'
)
```
