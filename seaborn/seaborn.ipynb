```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

plt.scatter(df[df.columns[0]],df[df.columns[3]])
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x=df.columns[0],
    y=df.columns[3],
)
plt.show()
```

```python
sns.barplot(
    data=df,
    x=df.columns[0],
    y=df.columns[3],
)
```

```python
sns.histplot(
    data=df,
    x=df.columns[3],
    kde=True
)
plt.show()
```
