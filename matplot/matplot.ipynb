```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.df = pd.read_csv("../pandas/data.csv")

plt.plot(df[df.columns[5]], df[df.columns[1]])
plt.xlabel("Price")
plt.ylabel("Name")
plt.show()
```

```python
plt.scatter(df[df.columns[0]],df[df.columns[1]])
plt.xlabel("Feature X") 
plt.ylabel("Feature Y")
plt.title("Scatter Plot")
plt.show()
```

```python
counts = df[df.columns[0]].value_counts()
plt.bar(counts.index, counts.values)
plt.xlabel("Category")
plt.ylabel("Count")
plt.title("Bar Plot")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df[df.columns[0]], bins=10)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(df[df.columns[0]], df[df.columns[1]])
plt.title("Line Plot")

plt.subplot(1,2,2)
plt.scatter(df[df.columns[0]], df[df.columns[1]])
plt.title("Scatter Plot")

plt.show()
```

```python
plt.plot(df[df.columns[0]], df[df.columns[1]],
         color='blue', linestyle='--', marker='o')
plt.grid(True)
plt.show()
```
