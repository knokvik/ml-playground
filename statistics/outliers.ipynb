```python
import numpy as np
```

```python
data = [
    22, 25, 28, 30, 32, 35, 40, 42, 45, 48, 
    50, 52, 55, 58, 60, 2, 5, 150, 200, 44, 
    41, 39, 37, 33, 31, 29, 27, 24, 21, 59, 
    56, 54, 51, 49, 46, 43, 38, 36, 34, 26
]
```

```python
data = np.sort(data)
data
```

```python
median = np.median(data)
median
```

```python
q1 = np.percentile(data,25)
q1
```

```python
q3 = np.percentile(data,75)
q3
```

```python
IQR = q3 - q1
IQR
```

```python
lower_bound  = q1 - (1.5*IQR)
lower_bound
```

```python
higher_bound = q3 + (1.5*IQR)
higher_bound
```

```python
l = []
for x in data :
    if x >= lower_bound and x <= higher_bound:
        l.append(x)
cleaned_data = np.array(l)
cleaned_data
```

```python
import seaborn as sns
sns.boxplot(x=cleaned_data)
```
