```python
import numpy as np

marks = np.array([
    [78, 85, 90],
    [60, 65, 70],
    [88, 92, 95],
    [50, 55, 58]
])
```


Tasks

Find average marks of each student

Find average marks of each subject

Find top scorer (row index)

Find students who scored more than 80 in all subjects

```python
print(marks.mean(axis=1))

print(marks.mean(axis=0))

avg = (marks.mean(axis=1))
print(np.argmax(avg))

high_score = np.where(np.all(marks > 80 , axis = 1))[0]
print(high_score)

```

```python
salaries = np.array([25000, 40000, 55000, 30000, 80000, 45000])
```

Tasks

Extract salaries above 40,000

Increase all salaries by 10%

Count how many employees earn more than average salary

Normalize salaries (mean = 0, std = 1)

```python
print(salaries[salaries > 40000])

print(salaries + ( salaries * .1))

avg = salaries.mean()
print(salaries[salaries > avg])

mean = salaries.mean()
std = salaries.std()
normalize = ( salaries - mean ) / std
print(normalize)
```

Tasks

Detect invalid values (-99)

Replace invalid values with mean of valid readings

Find min & max temperature

Standardize the cleaned sensor data

```python
sensor = np.array([22.5, 23.0, -99.0, 24.1, 23.8, -99.0, 25.0])
```

```python
invalid = sensor == -99
print(invalid)

valid_mean = sensor[~invalid].mean()
copy = valid_mean
sensor[invalid] = valid_mean
print(sensor)

sensor_standardized = (
    sensor - sensor.mean()
) / sensor.std()

print(sensor_standardized)
```
